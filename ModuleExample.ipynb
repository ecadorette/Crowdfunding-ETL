```python
#get the unique values in the category&sub-category column
crowdfunding_info_df['category & sub-category'].unique()
```

```python
#split the category&subcategory columns and add them in two new columns
crowdfunding_info_df[['category','subcategory']] = crowdfunding_info_df["category & sub-category"].str.split('/',n=1,expand=True)
crowdfunding_info_df.head(10)
```

```python
# Get the distinct values in the category and subcategory columns.
print(crowdfunding_info_df["category"].nunique())
print(crowdfunding_info_df["subcategory"].nunique())
```

```python
#get the unique categories and subcategories in separate lists
categories = crowdfunding_info_df["category"].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()
print(categories)
print()
print(subcategories)
```

```python
#we need to create unique numbers for categories and subcategories. so create two NumPy arrays that have unique
#numbers for the two new columns

#import the Numpy dependancy
import numpy as np

#create numpy arrays that have 1-10 for the category ids and 1-25 for the subcategoy ids
category_ids = np.arange(1,10)
subcategory_ids = np.arange(1,25)
print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat0" to each category_id. 
cat_ids = ["cat0" + str(cat_id) for cat_id in category_ids]

# Use a list comprehension to add "scat0" to each subcategory_id. 
scat_ids = ["scat0" + str(scat_id) for scat_id in subcategory_ids ]

print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the cat_ids array as the category_id and categories list as the category name.
category_df = pd.DataFrame({
    "category_id": cat_ids,
    "category" : categories
})

# Create a subcategory DataFrame with the scat_ids array as the subcategory_id and subcategories list as the subcategory name.
subcategory_df = pd.DataFrame({
    "subcategory_id": scat_ids,
    "subcategory" : subcategories
})
category_df
```

```python
subcategory_df.head(10)
```

```python
#export to csv files

#create the output file
output_data_file = "categories.csv"
#export
category_df.to_csv(output_data_file,index=False)
```

```python
#export to csv files

#create the output file
output_data_file2 = "subcategories.csv"
#export
subcategory_df.to_csv(output_data_file2,index=False)
```

```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df.
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
#rename blurb, launched_at, and deadline columns
campaign_df = campaign_df.rename(columns={'blurb':'description', 'launched_at':'launch_date', 'deadline':'end_date'})
campaign_df.head()
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df[["goal","pledged"]] = campaign_df[["goal","pledged"]].astype(float)
campaign_df.head()
```

```python
#looking for .dt.strftime('%Y-%m-%d') error below with 2020-02-13
#print(campaign_df[campaign_df["launch_date"].eq('2020-02-13')])
#print(campaign_df[campaign_df["end_date"].eq('2020-02-13')])
new_campaign_df = campaign_df[campaign_df.launch_date != '2020-02-13']
print(new_campaign_df[new_campaign_df["launch_date"].eq('2020-02-13')])
```

```python
# Format the launched_date and end_date columns to datetime format.
from datetime import datetime as dt
new_campaign_df["launch_date"] = pd.to_datetime(campaign_df["launch_date"], unit='s').dt.strftime('%Y-%m-%d') 
new_campaign_df["end_date"] = pd.to_datetime(campaign_df["end_date"], unit='s').dt.strftime('%Y-%m-%d')
#new_campaign_df.head()
print(new_campaign_df[new_campaign_df["launch_date"].eq('2020-02-13')])
```
